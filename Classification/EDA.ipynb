# 1. 查看赛方数据集分割比例

```python
import os

# Define the path to the directory
directory_path = '/root/CDataset/Real'

# Initialize counters
train_count = 0
val_count = 0
test_count = 0

# Iterate through all files in the directory
for filename in os.listdir(directory_path):
    # Check if the file is an image
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Count based on the filename
        if 'train' in filename:
            train_count += 1
        elif 'val' in filename:
            val_count += 1
        elif 'test' in filename:
            test_count += 1

# Calculate total and proportions
total_count = train_count + val_count + test_count
train_ratio = train_count / total_count if total_count else 0
val_ratio = val_count / total_count if total_count else 0
test_ratio = test_count / total_count if total_count else 0

# Print the results
print(f"Train images: {train_count}")
print(f"Validation images: {val_count}")
print(f"Test images: {test_count}")
print(f"Proportions - Train: {train_ratio:.2%}, Validation: {val_ratio:.2%}, Test: {test_ratio:.2%}")
```
